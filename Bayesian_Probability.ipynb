```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/IMDB Dataset.csv.zip')
```

```python
data.head()
```

```python
data.info()
```

```python
positive = (data['sentiment'] == "positive").mean()
print(positive)
```

```python
positive_data = data[data['sentiment'] != "negative"]
```

```python
positive_data.head(10)
```

```python
positive_data.info()
```

From a total dataset of 50000 values i dropped all the values that were negative leaving me with 25000 values probablity being total numbers of positive values over total number of values

```python
keywords = ["wonderful", "stunning", "amazing", "like"]
```

```python
wonderful = positive_data['review'].str.contains('wonderful', case=False).mean()
```

```python
print(wonderful)
```

```python
wonderful_margin = data['review'].str.contains('wonderful', case=False).mean()
print(wonderful_margin)
```

```python
wonderful_posterior = (wonderful * positive /wonderful_margin)
print(wonderful_posterior)
```

```python
stunning = positive_data['review'].str.contains('stunning', case=False).mean()
print(stunning)
```

```python
stunning_margin = data['review'].str.contains('stunning', case=False).mean()
print(stunning_margin)
```

```python
stunning_posterior = (stunning * positive /stunning_margin)
print(stunning_posterior)
```

```python
amazing = positive_data['review'].str.contains('amazing', case=False).mean()
print(amazing)
```

```python
amazing_margin = data['review'].str.contains('amazing', case=False).mean()
print(amazing_margin)
```

```python
amazing_posterior = (amazing * positive /amazing_margin)
print(amazing_posterior)
```

```python
like = positive_data['review'].str.contains('like', case=False).mean()
print(like)
```

```python
like_margin = data['review'].str.contains('like', case=False).mean()
print(like_margin)
```

```python
like_posterior = (like * positive /like_margin)
print(like_posterior)
```
